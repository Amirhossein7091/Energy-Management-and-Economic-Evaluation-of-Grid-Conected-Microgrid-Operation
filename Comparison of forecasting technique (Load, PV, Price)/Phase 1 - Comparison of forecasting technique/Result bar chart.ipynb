```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```
models = ['LSTM', 'Seq2Seq', 'AttSeq2Seq', 'TX', 'LSTM', 'Seq2Seq', 'AttSeq2Seq', 'TX']
loadMAE = [92.84, 85.37, 101.47, 84.13, 95.08, 87.88, 104.38, 88.04]
solarMAE = [236.05, 207.01, 344.13, 276.26, 245.97, 218.64, 345.35, 280.35]
priceMAE = [3.1962, 3.24, 5.78, 2.82, 3.46, 3.54, 6.13, 3.02]
types = ['training', 'training', 'training', 'training', 
         'testing', 'testing', 'testing', 'testing']

df = pd.DataFrame({'model': models, 'loadMAE': loadMAE, 'solarMAE': solarMAE,
                   'priceMAE': priceMAE, 'types': types})
#df.set_index('model', inplace=True, drop=True)
df.head(8)
```

```
plt.figure(figsize=(15, 6))
ax = sns.barplot(x='model', y='loadMAE', hue='types', data=df)
ax.set_ylabel('Load MAE (W)')
ax.legend(loc=3)
for p in ax.patches:
    #print ((p.get_x(), p.get_width(), p.get_height()))
    ax.annotate("%.4f" % p.get_height(), xy=(p.get_x()+p.get_width()/5, p.get_height()),
                xytext=(5, 5), textcoords='offset points', ha="left", va="center")
plt.title('Load MAE Corresponding to Different Forecasting Models')
plt.show()
```

```
plt.figure(figsize=(15, 6))
ax = sns.barplot(x='model', y='solarMAE', hue='types', data=df)
ax.set_ylabel('Solar MAE (W)')
ax.legend(loc=3)
for p in ax.patches:
    #print ((p.get_x(), p.get_width(), p.get_height()))
    ax.annotate("%.4f" % p.get_height(), xy=(p.get_x()+p.get_width()/5, p.get_height()),
                xytext=(5, 5), textcoords='offset points', ha="left", va="center")
  
plt.title('Solar MAE Corresponding to Different Forecasting Models')
plt.show()
```

```
plt.figure(figsize=(15, 6))
ax = sns.barplot(x='model', y='priceMAE', hue='types', data=df)
ax.set_ylabel('Price MAE ($)')
ax.legend(loc=3)
for p in ax.patches:
    #print ((p.get_x(), p.get_width(), p.get_height()))
    ax.annotate("%.4f" % p.get_height(), xy=(p.get_x()+p.get_width()/5, p.get_height()),
                xytext=(5, 5), textcoords='offset points', ha="left", va="center")
    
plt.title('Price MAE Corresponding to Different Forecasting Models')
plt.show()
```
